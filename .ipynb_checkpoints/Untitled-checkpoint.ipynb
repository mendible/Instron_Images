```python
import cv2
import numpy as np
import os
import ipywidgets as widgets
import matplotlib.pyplot as plt
```

```python
# set the names of the images that are the "calibration" images
first_im_filename = 'DSC_0297.JPG'
last_im_filename = 'example.jpg'
# activated_relaxed_filename = 'DSC_0430.JPG'
# bleached_filename = '.jpg'

#ignore those files
ignore_files = [last_im_filename] #, activated_relaxed_filename, bleached_filename]

# create the list of files that should be only the testing set
file_list = [f for f in os.listdir('3') if f.endswith('.JPG') and not f in ignore_files ]

file_list.sort()
```

```python
# show the images
for im_file in file_list:
    img = plt.imread(im_file)
    plt.imshow(img)
    plt.show()
```

```python
last_im_filename = 'example.jpg'
img = plt.imread(last_im_filename)
a = widgets.FloatRangeSlider(
    value=[5, 7.5],
    min=0,
    max=10.0,
    step=0.1,
    description='Test:',
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)


b = widgets.FloatRangeSlider(
    value=[5, 7.5],
    min=0,
    max=10.0,
    step=0.1,
    description='Test:',
    disabled=False,
    continuous_update=True,
    orientation='Vertical',
    readout=True,
    readout_format='.1f',
)


ui = widgets.HBox([a, b])
def f(a, b):
    plt.imshow(img)
    print((a, b))

out = widgets.interactive_output(f, {'a': a, 'b': b})

display(ui, out)
```

```python


for im_file in ignore_files: 
    img = plt.imread(im_file)
    m,n,k = img.shape
    
    widgets.IntRangeSlider(
    value=[1, m],
    min=1,
    max=m,
    step=1,
    description='Horizontal area:',
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='d',
    )
    
    widgets.IntRangeSlider(
    value=[1, n],
    min=1,
    max=n,
    step=1,
    description='Vertical Area:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='d',
    )
    plt.imshow(img)
```
